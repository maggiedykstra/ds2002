```python

# Allow the user to choose a module from the Quote Summary Endpoint.
module= input("Choose a module from the Quote Summary Endpoint: ") #must be summaryDetail to work, need this to get 52 week high and low

symbol = input("Choose a stock symbol: ")

apikey="NvydlJyKMH7p1TQIgHXUp1z8IGO55P6BaMJMVASy"


#Make an API request using the chosen module.

url = f"https://yfapi.net/v11/finance/quoteSummary/{symbol}?modules={module}&apiKey={apikey}"


querystring = {"modules": f"{module},financialData, summaryDetail",
               "symbol": symbol}

headers = {
    'x-api-key': apikey
    }

# response = requests.get(url, headers=headers, params=querystring)



response = requests.get(url, headers=headers, params=querystring)
data = response.json()["quoteSummary"]["result"][0]
# print(data)


summary_detail = data.get("summaryDetail", {})
financial_data = data.get("financialData", {})


# Find:
# 52 Week High
# 52 Week Low
# Return on Assets (ROA)

stock_info = {
  "52 Week High": summary_detail.get("fiftyTwoWeekHigh", {}).get("fmt", "N/A"),
  "52 Week Low": summary_detail.get("fiftyTwoWeekLow", {}).get("fmt", "N/A"),
  "Return on Assets (ROA)": financial_data.get("returnOnAssets", {}).get("fmt", "N/A"),
}

#  - Convert the response into a Pandas DataFrame.
#  - Display the DataFrame.
df = pd.DataFrame([stock_info])
df






```

```python

region = input("Region: ")

apikey="NvydlJyKMH7p1TQIgHXUp1z8IGO55P6BaMJMVASy"

url = f"https://yfapi.net/v1/finance/trending/{region}"

querystring = {"region":region}

headers = {
    'x-api-key': apikey
    }

trending_tickers = requests.get(url, headers=headers, params=querystring)
trending_data = trending_tickers.json()
trending_stocks = trending_data.get("finance", {}).get("result", [{}])[0].get("quotes", [])
# print(trending_stocks)
tickers = []
for stock in trending_stocks:
    tickers.append(stock["symbol"])
# print(tickers)

stock_data = []

for ticker in tickers:
    quote_url = f"https://yfapi.net/v11/finance/quoteSummary/{ticker}"
    querystring = {"modules": "summaryDetail,price", "apikey": apikey}

    response = requests.get(quote_url, headers=headers, params=querystring)
    stock_info = response.json().get("quoteSummary", {}).get("result", [{}])[0]

    price_info = stock_info.get("price", {})
    summary_info = stock_info.get("summaryDetail", {})

    stock_entry = {
        "Stock Name": price_info.get("shortName", "N/A"),
        "Ticker": ticker,
        "Current Price": price_info.get("regularMarketPrice", {}).get("raw", "N/A"),
        "52 Week High": summary_info.get("fiftyTwoWeekHigh", {}).get("raw", "N/A"),
        "52 Week Low": summary_info.get("fiftyTwoWeekLow", {}).get("raw", "N/A"),
    }

    stock_data.append(stock_entry)

df = pd.DataFrame(stock_data)
df


```
